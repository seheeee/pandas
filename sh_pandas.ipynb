## 함수
- map함수 : DataFrame 타입이 아니라, 반드시 Series 타입에서만 사용
- apply함수 : 커스텀 함수를 사용하기 위해 DataFrame에서 복수 개의 컬럼이 필요하다면, apply함수를 사용
- applymap함수 : DataFrame클래스의 함수이긴 하나, 각 요소(element)별로 작동. 벡터에 스칼라를 연산하면, 벡터의 요소 하나하나에 해당 연산을 해주는 것과 같음(elementwise). applymap에 인자로 전달하는 커스텀함수가 Single value로부터 Single value를 반환한다는 점이 중요하다

[pipe vs applymap vs apply]  
pipe와 applymap과의 차이는 반환하는 결과가 pipe가 더 다양하다는 것으로,<br>
applymap은 데이터프레임 원소에 적용하므로 데이터프레임으로만 리턴되지만,<br>
pipe는 사용자 함수의 리턴값에 따라 개별 값, 시리즈, 데이터프레임을 반환할 수 있다.<br>  
현재 가장 많이 쓰이는 메서드는 apply인데,<br>
그 이유는 다양한 형태로 값을 입력 받을 수 있고, 반환하는 형태도 폭이 넓기 때문.<br>
결과는 같아보이지만, 시리즈를 이용해 데이터프레임의 행과 열도 이용할 수 있으며 시리즈 그 자체도 함수로 활용할 수 있다. <br>

만약 데이터프레임에 함수를 적용해야한다면 리턴하는 형태가 많은 pipe 사용도 고려

````python
#범주화
def get_category(x): #여기서 x는 후에 적용할 data의 원소들을 뜻함
    cat = ''
    if x < 20:
        cat = 'child'
    elif x >= 20:
        cat = 'adult'
    return cat

df['age_class'] = df.age.apply(get_category) #df.age에 함수를 적용하여 'age_class'에 넣겠다.
df[['age','age_class']].head()
````

````python
#소문자를 대문자로 바꾸는 함수를 만들고 df.age_class에 적용하세요.
def up(x):
    return x.upper()

df.age_class = df.age_class.map(up) #df.age_class가 sr이므로 map적용해도 OK
df.head()
````

### 파이썬 : time, datetime

````python
#timestamp > str
t = time.time()
print(t, type(t))

s = str(t)
print(s, type(s))
````

````python
#datetime > str
t = datetime.now()
print(t, type(t))

s = datetime.strftime(t, '%Y-%m-%d %H:%M:%S')
print(s, type(s))

#str > datetime
s = '2022-07-03 11:22:43'
t = datetime.strptime(s,'%Y-%m-%d %H:%M:%S')
print(t, type(t))
````

````python
#datetime > timestamp : timetuple()로 바꿔줘야함.
s = '2022-07-03 11:22:43'
dt = datetime.strptime(s, '%Y-%m-%d %H:%M:%S')
ts = time.mktime(dt.timetuple()) # time사용
print(ts,type(ts))

#timestamp > datetime
ts = time.time()
print(ts,type(ts))
dt = datetime.fromtimestamp(ts)
print(dt,type(dt))
````

## 데이터프레임 결합
- pd.concat, pd.merge, join
* join()메소드는 pd.merge() 함수를 기반으로 생성 > 기본 작동 방식이 서로 비슷
<br>

* join() : "행 인덱스" 기준 결합, on=keys 옵션 설정하면 열을 기준으로 결합 가능
* pd.merge() : 열 인덱스 기준 결합 


````python
data1 = np.arange(1,21).reshape(4,5)
data2 =np.arange(11,31).reshape(5,4)
df1 = pd.DataFrame(data1, columns = list('abcde'))
df2 = pd.DataFrame(data2, columns = list('dehi'))
print(df1, '\n')
print(df2)
````

### join
- df1.join(df2, how = 'inner')
- 행인덱스 기준 결합
- df1.join(df2, lsuffix='_a', rsuffix = '_b') : col겹칠 때 구분

````python
#how = 'left'가 기본옵션 : df1의 행인덱스를 기준으로 결합
df3 = df1.join(df2, lsuffix='_df1', rsuffix='_df2')
df3
````

````python
df1.join(df2, lsuffix='_df1', rsuffix = '_df2', how='inner')
df1.join(df2, lsuffix='_df1', rsuffix = '_df2', how='outer')
````

- 행 index 기준으로 붙는다! 의 뜻

````python
#df1, df2를 join하는데, 인덱스를 ind로 설정
df1['ind'] = [1,2,3,4]
df2['ind'] = [1,2,3,4,5]

df1.set_index('ind', inplace=True)
df2.set_index('ind', inplace=True)
df1, df2
````

````python
#how=inner로 했기 때문에, df1의 ind까지만 출력
df1.join(df2, lsuffix='_df1', rsuffix='_df2', how='inner' )

#how=outer로 했기 때문에, df2의 ind까지 모두 출력
df1.join(df2, lsuffix='_df1', rsuffix='_df2', how='outer' )
````

### merge
* merge() 함수는 sql의 join 명령과 비슷한 방식으로 어떤 기준에 의해 두 데이터프레임을 병합하는 개념
* 기준이 되는 열이나 인덱스를 키라고 부름
* 키가 되는 열이나 인덱스는 반드시 양쪽 데이터프레임에 모두 존재함
<br>

- pd.merge(left_data, right_data, how='inner', on= 기준열) # default
- 'how' option : inner(교집합), outer(합집합), left(왼쪽 데이터기준), right(오른쪽 데이터 기준)

````python
data1 = np.arange(1,21).reshape(4,5)
data2 =np.arange(11,31).reshape(5,4)
df1 = pd.DataFrame(data1, columns = list('abcde'))
df2 = pd.DataFrame(data2, columns = list('dehi'))

df1['ind'] = [1,2,3,4]
df2['ind'] = [1,2,3,4,5]

print(df1, '\n')
print(df2)
````

````python
pd.merge(df1,df2, on='ind', how='outer')
````

## scaling


### MinMaxScaler(정규화)
- 모든 값을 0~1로 정규화

````python
from sklearn.datasets import load_boston #data가져오기
boston = load_boston() #주택가격 예측모델
print(boston.keys())
print(boston.feature_names) #주택가격에 영향을 주는 독립변수들
````

````python
df_boston = pd.DataFrame(boston.data, columns = boston.feature_names)
df_boston['target'] = boston.target
df_boston.head()
````

````python
#RM에 대한 MinMaxScaling
from sklearn.preprocessing import MinMaxScaler #클래스
scaler = MinMaxScaler() #객체에 할당

df_boston['RM'] = scaler.fit_transform(df_boston[['RM']])
#fit_transform : fit (준비) transform(시작) > 따로해도 됨.
df_boston[['RM']].describe() #min =0, max =1
````

### StandardScaler(정규화)

````python
from sklearn.preprocessing import StandardScaler #클래스임
scaler = StandardScaler() #객체에 할당

df_boston['INDUS'] = scaler.fit_transform(df_boston[['INDUS']])
df_boston[['INDUS']].describe() #std=0 mean=1
````
