```python
import seaborn as sns

#Seaborn 데이터셋 목록 (기본제공)
sns.get_dataset_names()
```

```python
df = sns.load_dataset('titanic') 
df.head()
```

```python
df.age.value_counts() #value count
```

```python
# 범주화

def get_category(x):
    cat = ''
    if x < 20:
        cat = 'child'
    elif x < 30:
        cat = 'Young Adult'
    elif x < 50:
        cat = 'Adult'
    else:
        cat = 'Elderly'
    return cat


df['age_class'] = df.age.apply(lambda x:get_category(x))
df[['age','age_class']].head()
```

```python
# 소문자를 대문자로 변환
def upper_(x):
    return x.upper()
df['sex'] = df.sex.apply(lambda x :upper_(x))
df[['sex']].head() #대괄호 하나와 두개의 차이는?
```
